## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print("Total pages:", len(df))

zero_click_pct = (df["clicks_90d"] == 0).mean() * 100
print(f"Pages with impressions but 0 clicks (90d): {zero_click_pct:.1f}%")

page1_median_ctr = df[df["position_tier"] == "page_1"]["ctr"].median()
page1_count = (df["position_tier"] == "page_1").sum()
print(f"Median CTR for 'page_1' tier ({page1_count} pages): {page1_median_ctr:.2f}%")
```
